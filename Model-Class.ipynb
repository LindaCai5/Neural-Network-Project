```python
#import libs
import h5py
import numpy as np 
from matplotlib import pyplot as plt
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
import csv
import pandas as pd
```

```python
#Create Model class
class MultiDigitModel:
    
    def __init__(self):
        #first digit model
        self.model_1 = Sequential()
        self.model_1.add(Conv2D(16, kernel_size=(5,5), activation='relu', input_shape=(64,64,1)))
        self.model_1.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_1.add(Conv2D(32, (5, 5), activation='relu'))
        self.model_1.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_1.add(Conv2D(64, (3, 3), activation='relu'))
        self.model_1.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_1.add(Dropout(0.25))
        self.model_1.add(Flatten())
        self.model_1.add(Dense(128, activation="relu"))
        self.model_1.add(Dropout(0.5))
        self.model_1.add(Dense(10,activation="softmax"))
        #second digit model
        self.model_2 = Sequential()
        self.model_2.add(Conv2D(16, kernel_size=(5,5), activation='relu', input_shape=(64,64,1)))
        self.model_2.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_2.add(Conv2D(32, (5, 5), activation='relu'))
        self.model_2.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_2.add(Conv2D(64, (3, 3), activation='relu'))
        self.model_2.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_2.add(Dropout(0.25))
        self.model_2.add(Flatten())
        self.model_2.add(Dense(128, activation="relu"))
        self.model_2.add(Dropout(0.5))
        self.model_2.add(Dense(11,activation="softmax"))
        #third digit model
        self.model_3 = Sequential()
        self.model_3.add(Conv2D(16, kernel_size=(5,5), activation='relu', input_shape=(64,64,1)))
        self.model_3.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_3.add(Conv2D(32, (5, 5), activation='relu'))
        self.model_3.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_3.add(Conv2D(64, (3, 3), activation='relu'))
        self.model_3.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_3.add(Dropout(0.25))
        self.model_3.add(Flatten())
        self.model_3.add(Dense(128, activation="relu"))
        self.model_3.add(Dropout(0.55))
        self.model_3.add(Dense(11,activation="softmax"))
        #fourth digit model
        self.model_4 = Sequential()
        self.model_4.add(Conv2D(16, kernel_size=(5,5), activation='relu',  input_shape=(64,64,1)))
        self.model_4.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_4.add(Conv2D(32, (5, 5), activation='relu'))
        self.model_4.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_4.add(Conv2D(64, (3, 3), activation='relu'))
        self.model_4.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_4.add(Dropout(0.25))
        self.model_4.add(Flatten())
        self.model_4.add(Dense(128, activation="relu"))
        self.model_4.add(Dropout(0.6))
        self.model_4.add(Dense(11,activation="softmax"))
        #fifth digit model
        self.model_5 = Sequential()
        self.model_5.add(Conv2D(16, kernel_size=(5,5), activation='relu',  input_shape=(64,64,1)))
        self.model_5.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_5.add(Conv2D(32, (5, 5), activation='relu'))
        self.model_5.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_5.add(Conv2D(64, (3, 3), activation='relu'))
        self.model_5.add(MaxPooling2D(pool_size=(2, 2)))
        self.model_5.add(Dropout(0.25))
        self.model_5.add(Flatten())
        self.model_5.add(Dense(128, activation="relu"))
        self.model_5.add(Dropout(0.6))
        self.model_5.add(Dense(11,activation="softmax"))

    
    
    #Fit data function
    def fit(self,X_train,y_train, num_epoch = 15):
        X_train = X_train.astype('float32')
        X_train /= 255
        history = []
        #opt = keras.optimizers.Adam(learning_rate=0.001)
        print("===========================================================")
        print("First Model")
        print("===========================================================")
        self.model_1.compile(loss='sparse_categorical_crossentropy', optimizer="adam", metrics=['accuracy'])
        hist1 = self.model_1.fit(x=X_train,y=y_train[:,0], epochs=num_epoch, validation_split = 0.1)
        history.append(hist1.history.get('accuracy')[-1])
        print("===========================================================")
        print("Second Model")
        print("===========================================================")
        self.model_2.compile(loss='sparse_categorical_crossentropy', optimizer="adam", metrics=['accuracy'])
        hist2 = self.model_2.fit(x=X_train,y=y_train[:,1], epochs=num_epoch, validation_split = 0.1)
        history.append(hist2.history.get('accuracy')[-1])
        print("===========================================================")
        print("Third Model")
        print("===========================================================")
        self.model_3.compile(loss='sparse_categorical_crossentropy', optimizer="adam", metrics=['accuracy'])
        hist3 = self.model_3.fit(x=X_train,y=y_train[:,2], epochs=num_epoch, validation_split = 0.1)
        history.append(hist3.history.get('accuracy')[-1])
        print("===========================================================")
        print("Fourth Model")
        print("===========================================================")
        self.model_4.compile(loss='sparse_categorical_crossentropy', optimizer="adam", metrics=['accuracy'])
        hist4 = self.model_4.fit(x=X_train,y=y_train[:,3], epochs=num_epoch, validation_split = 0.1)
        history.append(hist4.history.get('accuracy')[-1])
        print("===========================================================")
        print("Fifth Model")
        print("===========================================================")
        self.model_5.compile(loss='sparse_categorical_crossentropy', optimizer="adam", metrics=['accuracy'])
        hist5 = self.model_5.fit(x=X_train,y=y_train[:,4], epochs=num_epoch, validation_split = 0.1)
        history.append(hist5.history.get('accuracy')[-1])
        return history
    
    #Evaluate model, take the prodaut and average of the 5 models
    def evaluate(self,X_test,y_test):
        X_test= X_test.astype('float32')
        X_test /= 255
        score1 = self.model_1.evaluate(X_test, y_test[:,0])
        print("The first digit score:", score1)
        score2 = self.model_2.evaluate(X_test, y_test[:,1])
        print("The second digit score:", score2)
        score3 = self.model_3.evaluate(X_test, y_test[:,2])
        print("The third digit score:", score3)
        score4 = self.model_4.evaluate(X_test, y_test[:,3])
        print("The fourth digit score:", score4)
        score5 = self.model_5.evaluate(X_test, y_test[:,4])
        print("The fifth digit score:", score5)
        average = [(score1[0]+score2[0]+score3[0]+score4[0]+score5[0])/5.0, score1[1]*score2[1]*score3[1]*score4[1]*score5[1]]
        print("The average score:", average)
        return average
    
    #Predit data function   
    def predict(self,X_test):
        X_test = X_test.astype('float32')
        X_test /= 255
        digit1 = self.model_1.predict(X_test)
        digit1 = [x.argmax() for x in digit1]
        digit2 = self.model_2.predict(X_test)
        digit2 = [x.argmax() for x in digit2]
        digit3 = self.model_3.predict(X_test)
        digit3 = [x.argmax() for x in digit3]
        digit4 = self.model_4.predict(X_test)
        digit4 = [x.argmax() for x in digit4]
        digit5 = self.model_5.predict(X_test)
        digit5 = [x.argmax() for x in digit5]
        result = [str(digit1[i])+str(digit2[i])+str(digit3[i])+str(digit4[i])+str(digit5[i]) for i in range(len(digit1))]
        return result
```

```python
#Preprocess
f1 = h5py.File('synthetic.h5', 'r+')
#print(list(f1.keys()))
X_train = f1['train_dataset']
X_test = f1['test_dataset']
Y_train = f1['train_labels']
dxtrain= np.array(X_train[()])
dxtest= np.array(X_test[()])
dytrain= np.array(Y_train[()])
x_train, x_test, y_train, y_test = train_test_split(dxtrain,dytrain, test_size=0.2, random_state=42)
input_shape = (64, 64, 1)
#if you want to see a image
#plt.imshow(dxtrain[image_index].reshape(64, 64)) 
#if you want ot check data shape
# print (dxtrain.shape)
# print (dytrain.shape)
# print (dxtest.shape)
```

```python
#Fit data and Evaluate
Model = MultiDigitModel()
rowNum=0
#create a 5 column by 3 row array, where each row stores the number of epoc
#goes from epochs = 5 , 10, 15, 20, 25, 30
epochArray = np.zeros(shape=(6,3))
i=5
rowNum = 0
while i<=30:
    hist = Model.fit(x_train, y_train, 5)  #change to 5 later
    acc = Model.evaluate(x_test, y_test)
    epochArray[rowNum, 0] = i
    tmp = 1.0
    for x in hist:
        tmp *= x
    epochArray[rowNum, 1] = tmp
    epochArray[rowNum, 2] = acc[1]
    rowNum += 1
    i += 5



    
```
